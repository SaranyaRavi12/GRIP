# Data Visualisation

```python
plt.figure(figsize=(16,8))
plt.bar("Sub-Category","Category",data=sample_drop, color= 'Orange')
plt.title('Category Vs Sub-Category',fontsize=25)
plt.xlabel('Sub-Category',fontsize=20)
plt.ylabel('Category ',fontsize=20)
plt.xticks(rotation=45)
plt.show()
```

```python
sample_drop.hist(bins=50,figsize=(25,20))
plt.show()
```

```python
# Count the repeatable states
sample_drop['State'].value_counts() 
```

```python
plt.figure(figsize=(18,18))

sns.countplot(x=sample_drop['State'])
plt.title("STATE")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(14,12))
sample_drop['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

```python
sample_drop.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar()
plt.title('TOTAL PROFIT AND SALES PER SUB-CATEGORY')
plt.show()

sns.set(style="whitegrid")
plt.figure(2,figsize=(16,8))

sns.barplot(x='Sub-Category',y='Profit',data=sample_dataset, palette='Spectral')
plt.suptitle('PIE CONSUMPTION PATTERN IN THE UNITED STATES',fontsize=25)

plt.show()
```

```python
# Analysis based on Shipping mode
sns.countplot(x=sample_dataset['Ship Mode'])
```

```python
# ploting pair plot for Sub-Category
figsize=(20,15)
sns.pairplot(sample_drop,hue='Sub-Category')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.lineplot('Discount','Profit',data=sample_drop,color='y', label= 'Discount')
plt.legend()
plt.show()
```
